$$
\begin{split}
& \underline{\text{Hyperbolic Cosine Function}}\text{:} \\ \\ 
& \quad \quad \quad \cosh\big(x\big) = \frac{1}{2}\Big(e^{x} + e^{-x}\Big) \\ \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Section A : Inverse Hyperbolic Cosine via Quadratic Formula}} \\ \\
& \quad \quad \quad \quad \space \space \cosh\big(x\big) = y \space \space \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \big(\text{function definition}\big) \\ \\
&  \quad \quad \quad \quad \quad \quad \quad \quad = \frac{1}{2}\bigg(e^{x} + e^{-x}\bigg) \\ \\
&  \quad \quad \quad \quad \quad \Rightarrow \quad y = \frac{e^{2x} + 1}{2e^{x}} \quad \quad \quad  \quad \quad \quad \quad \quad \quad \quad \quad \space \space \quad \quad \quad \quad \quad \big(\text{rewrite into fraction}\big)\\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad 2ye^{x} = e^{2x} + 1 \\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad 0 = e^{2x} - 2ye^{x} + 1 \quad \quad \quad  \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space \big(\text{rewrite fraction into quadratic equation equal to } 0\big) \ \\ \\
& \quad \quad \quad \quad  \quad \Rightarrow \quad 0 = \big(e^{x} - y\big)^{2} - \big(y^{2} - 1\big) \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \big(\text{complete the square}\big) \\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad e^{x} = y \pm \sqrt{y^{2} - 1} \quad \quad \quad \quad \quad \quad \quad \quad \space \space \quad \quad \quad \quad \quad \space \space \big(\text{isolate } e^{x} \text{ via quadratic solutions}\big) \\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad x = \ln\bigg[y \pm \sqrt{y^{2} - 1}\bigg] \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \big(\text{isolate } x \text{ from } e^{x}\big) \\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad x = \ln\bigg[y + \sqrt{y^{2} - 1}\bigg] \space, \quad \space \space \ln\bigg[y - \sqrt{y^{2} - 1}\bigg] \\ \\
&  \quad \quad \quad \quad \quad \Rightarrow \quad x = \ln\bigg[y + \sqrt{y^{2} - 1}\bigg] \space, \quad \space \space \ln\Bigg[\frac{1}{y + \sqrt{y^{2} - 1}}\Bigg] \quad \quad \space \big(\text{see Appendix A}\big) \\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad x = \ln\bigg[y + \sqrt{y^{2} - 1}\bigg] \space, \quad -\ln\bigg[y + \sqrt{y^{2} - 1}\bigg] \\ \\
& \quad \quad \quad  \quad \quad \Rightarrow \quad x = \operatorname{sgn\big(x\big)}\ln\bigg[y + \sqrt{y^{2} - 1}\bigg] \\ \\
&  \quad \quad \quad \quad \quad \Rightarrow \quad x = \ln\Bigg[\bigg(y + \sqrt{y^{2} - 1}\bigg)^{\operatorname{sgn(x)}}\Bigg] \\ \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Section B: Inverse Hyperbolic Cosine of Natural Logarithm}} \\ \\
& \quad \quad \quad \cosh\Big(\ln\big[x\big]\Big) = y \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \big(\text{function definition}\big) \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = \frac{1}{2}\Big(e^{\ln[x]} + e^{-\ln[x]}\Big) \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = \frac{1}{2}\Big(e^{\ln[x]} + e^{\ln\big[\frac{1}{x}\big]}\Big) \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = \frac{1}{2}\bigg(x + \frac{1}{x}\bigg) \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = \frac{x^{2} + 1}{2x} \\ \\
& \quad \quad \quad \space \space \space \Rightarrow \quad \ln\big[x\big] = \ln\Bigg[\bigg(y + \sqrt{y^{2} - 1}\bigg)^{\operatorname{sgn\big(\ln[x]\big)}}\Bigg] \quad \quad \quad \big(\text{see Section A}\big) \\ \\
& \quad \quad \quad \space \space \space \Rightarrow \quad x = \bigg(y + \sqrt{y^{2} - 1}\bigg)^{\operatorname{sgn\big(\ln[x]\big)}} \\ \\
& \quad \quad \quad \space \space \space \Rightarrow \quad x = \bigg(y + \sqrt{y^{2} - 1}\bigg)^{\operatorname{sgn(x - 1)}} \quad \quad \quad  \quad \quad \quad \quad \space \space \space \big(\text{see Appendix B}\big) \\ \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Appendix A : Relationship between two quadratic solutions}} \\ \\
& \quad \quad \quad \Big(x + \sqrt{x^{2} - 1}\Big)\Big(x - \sqrt{x^{2} - 1}\Big) = x^{2} + \Big(x\Big)\Big(\sqrt{x^{2} - 1}\Big) - \Big(x\Big)\Big(\sqrt{x^{2} - 1}\Big) + \Big(\sqrt{x^{2} - 1}\Big)\Big(-\sqrt{x^{2} - 1}\Big) \\ \\
& \quad \quad \quad \ \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = x^{2} + \Big(x\Big)\Big(\sqrt{x^{2} - 1}\Big) - \Big(x\Big)\Big(\sqrt{x^{2} - 1}\Big) - \Big(x^{2} - 1\Big) \\ \\
& \quad \quad \quad \ \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = x^{2} - \Big(x^{2} - 1\Big) \\ \\
& \quad \quad \quad \ \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = 1 \\ \\
& \quad \quad \quad \ \Rightarrow \quad x - \sqrt{x^{2} - 1} = \frac{1}{x + \sqrt{x^{2} - 1}} \\ \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Appendix B : Conversion of domain boundary between two separate solutions}} \\ \\
& \quad \quad \operatorname{sgn}\big(\ln[x]\big) = \begin{cases} \quad \space 1 \space , \quad \ln[x] \gt 0  \\ \\  \space -1 \space , \quad \ln[x] \lt 0\end{cases} \\ \\
& \quad \quad \quad \quad \quad \quad \space \space = \begin{cases} \quad \space 1 \space , \quad x \gt e^{0}  \\ \\  \space -1 \space , \quad x \lt e^{0}\end{cases} \\ \\
& \quad \quad \quad \quad \quad \quad \space \space = \begin{cases} \quad \space 1 \space , \quad x \gt 1  \\ \\  \space -1 \space , \quad x \lt 1\end{cases} \\ \\
& \quad \quad \quad \quad \quad \quad \space \space = \begin{cases} \quad \space 1 \space , \quad x - 1 \gt 0 \\ \\  \space -1 \space , \quad x - 1\lt 0 \end{cases} \\ \\
& \quad \quad \quad \quad \quad \quad \space \space = \operatorname{sgn}(x - 1)
\end{split}
$$
