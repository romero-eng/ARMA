$$
\begin{split}
& \underline{\text{Summary : DTFT with single complex root}} \\ \\
& \quad \quad \quad f(z) = 1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)z^{-1} \\ \\
&  \quad \Rightarrow \quad \Big|f\big(e^{j\omega}\big)\Big| = \sqrt{2|{\rho}|\Big[\Gamma + \cos\big(\omega \mp {\angle}{\rho}\big)\Big]} \space \space, \quad |{\rho}| = \bigg(\Gamma + \sqrt{\Gamma^{2} - 1}\bigg)^{\operatorname{sgn\big(|{\rho}| - 1\big)}} \\ \\
&  \quad \Rightarrow \quad {\angle}f\big(e^{j\omega}\big) = -\tan^{-1}\begin{pmatrix}\frac{\sin\big(\omega \mp {\angle}{\rho}\big)}{|{\rho}|^{-1} + \cos\big(\omega \mp {\angle}{\rho}\big)}\end{pmatrix} \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Section A : Rewriting DTFT into Cartesian Components}} \\ \\
& \quad \quad 1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega} = 1 + |{\rho}|e^{{\pm}j{\angle}{\rho} - j\omega} \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = 1 + |{\rho}|e^{-j(\omega \mp {\angle}{\rho})} \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = 1 + |{\rho}|\bigg[\cos\Big(-(\omega \mp {\angle}{\rho})\Big) + j\sin\Big(-(\omega \mp {\angle}{\rho})\Big)\bigg] \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = 1 + |{\rho}|\bigg[\cos\big(\omega \mp {\angle}{\rho}\big) - j\sin\big(\omega \mp {\angle}{\rho}\big)\bigg] \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = 1 + |{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) - j|{\rho}|\sin\big(\omega \mp {\angle}{\rho}\big) \\ \\
& \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = \bigg[1 + |{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big)\bigg] + j\bigg[-|{\rho}|\sin\big(\omega \mp {\angle}{\rho}\big)\bigg] \\ \\ \\
& \quad \quad\Rightarrow \quad \operatorname{Re}\bigg\{1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg\} \space = \space 1 + |{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) \\ \\
& \quad \quad \Rightarrow \quad \operatorname{Im}\bigg\{1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg\} \space = \quad -|{\rho}|\sin\big(\omega \mp {\angle}{\rho}\big) \\ \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Section B : Squared Magnitude of DTFT}} \\ \\
& \space \quad \quad \bigg|1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg|^{2} = \operatorname{Re}\bigg\{1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg\}^{2} + \operatorname{Im}\bigg\{1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg\}^{2} \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = \bigg[1 + |{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big)\bigg]^{2} + \bigg[-|{\rho}|\sin\big(\omega \mp {\angle}{\rho}\big)\bigg]^{2} \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = \bigg[1 + 2|{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) + |{\rho}|^{2}\cos^{2}\big(\omega \mp {\angle}{\rho}\big)\bigg] + \bigg[|{\rho}|^{2}\sin^{2}\big(\omega \mp {\angle}{\rho}\big)\bigg] \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = 1 + 2|{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) + |{\rho}|^{2}\cos^{2}\big(\omega \mp {\angle}{\rho}\big) + |{\rho}|^{2}\sin^{2}\big(\omega \mp {\angle}{\rho}\big) \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = 1 + 2|{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) + |{\rho}|^{2}\bigg[\cos^{2}\big(\omega \mp {\angle}{\rho}\big) + \sin^{2}\big(\omega \mp {\angle}{\rho}\big)\bigg] \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = 1 + 2|{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) + |{\rho}|^{2} \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = |{\rho}|^{2} + 1 + 2|{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big) \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = 2|{\rho}|\Bigg[\bigg(\frac{|{\rho}|^{2} + 1}{2|{\rho}|}\bigg) + \cos\big(\omega \mp {\angle}{\rho}\big)\Bigg] \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = 2|{\rho}|\bigg[\cosh\Big(\ln\big[|\rho|\big]\Big) + \cos\big(\omega \mp {\angle}{\rho}\big)\bigg] \quad \quad \Big(\text{see }\underline{\text{Section B}}\text{ in }\underline{\bf{\text{Inverse Hyperbolic Cosine}}}\Big) \\ \\
& \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \space \space \space = 2|{\rho}|\Big[\Gamma + \cos\big(\omega \mp {\angle}{\rho}\big)\Big] \space \space, \quad |{\rho}| = \bigg(\Gamma + \sqrt{\Gamma^{2} - 1}\bigg)^{\operatorname{sgn\big(|{\rho}| - 1\big)}} \\ \\ \\
& \Rightarrow \quad \bigg|1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg| = \sqrt{2|{\rho}|\Big[\Gamma + \cos\big(\omega \mp {\angle}{\rho}\big)\Big]} \space \space, \quad |{\rho}| = \bigg(\Gamma + \sqrt{\Gamma^{2} - 1}\bigg)^{\operatorname{sgn\big(|{\rho}| - 1\big)}} \\ \\ \\
\end{split}
$$

$$
\begin{split}
& \underline{\text{Section C : Phase of DTFT}} \\ \\
& \space \space \quad \quad {\angle}\bigg(1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg) = \space \space \space \tan^{-1}\begin{pmatrix}\frac{\operatorname{Im}\Big\{1 + \big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\big)e^{-j\omega}\Big\}}{\operatorname{Re}\Big\{1 + \big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\big)e^{-j\omega}\Big\}}\end{pmatrix} \\ \\
& \space \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = \space \space \space \tan^{-1}\begin{pmatrix}\frac{-|{\rho}|\sin\big(\omega \mp {\angle}{\rho}\big)}{1 + |{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big)}\end{pmatrix} \\ \\
& \space \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = -\tan^{-1}\begin{pmatrix}\frac{|{\rho}|\sin\big(\omega \mp {\angle}{\rho}\big)}{1 + |{\rho}|\cos\big(\omega \mp {\angle}{\rho}\big)}\end{pmatrix} \\ \\
& \space \space \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad \quad = -\tan^{-1}\begin{pmatrix}\frac{\sin\big(\omega \mp {\angle}{\rho}\big)}{|{\rho}|^{-1} + \cos\big(\omega \mp {\angle}{\rho}\big)}\end{pmatrix} \\ \\ \\
& \Rightarrow \quad {\angle}\bigg(1 + \Big(|{\rho}|e^{{\pm}j{\angle}{\rho}}\Big)e^{-j\omega}\bigg) = -\tan^{-1}\begin{pmatrix}\frac{\sin\big(\omega \mp {\angle}{\rho}\big)}{|{\rho}|^{-1} + \cos\big(\omega \mp {\angle}{\rho}\big)}\end{pmatrix} \\ \\ \\
\end{split}
$$
